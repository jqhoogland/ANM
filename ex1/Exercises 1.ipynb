#  Problem 1.2 Numerical Integration

```python
import numpy as np
import matplotlib.pyplot as plt
#import integrate_rect, integrate_trap, integrate_simpsons, error_scaling from .integrators
```

```python
def integrate(fn, a, b, n_steps):
    _sum = 0
    
    x_range = np.arange(a, b, (b-a)/n_steps)
    
    for x in x_range:
       _sum += fn(x)
    
    figure = plt.figure()
    ax = figure.add_subplot(1, 1, 1)
    ax.plot(x_range, fn(x_range))
    ax.set_title("Integration of {} with {} integration steps".format(fn, n_steps))
    return _sum / n_steps
```

```python
# 1.
exp_integration_10 = integrate(np.exp, -1., 1., 10)
exp_integration_100 = integrate(np.exp, -1., 1., 100)
exp_integration_1000 = integrate(np.exp, -1., 1., 1000)
exp_integration_10000 = integrate(np.exp, -1., 1., 10000)

print("Approx. int exp(x) over [-1, 1] 10 ^ {1, 2, 3, 4} integration steps (in order):",
      exp_integration_10, 
      exp_integration_100, 
      exp_integration_1000, 
      exp_integration_10000)

figure = plt.figure()

plt.plot([1, 2, 3, 4], [exp_integration_10, 
      exp_integration_100, 
      exp_integration_1000, 
      exp_integration_10000])
```
